```python
import polygon_logic
import ult
import pandas as pd
import numpy as np
import mapkit as mk
```

```python
# Creating DataFrame that can be ingested to create an ultindx
```

```python
data = polygon_logic.make_polygon_table('states.kml',idfield='NAME')
```

```python
# The output table looks something like this
```

```python
data
```

```python
# Now to create the ultindex.. this process is compute intensive so you can pass in a spark context to parallelize this procss
```

```python
ult.make_polygon_index(data,'states.h5')
```

```python
# Reading ultindex into memory
```

```python
index = ult.readh('states.h5')
```

```python
# Making a random dataset
```

```python
extrema = {'n': 50.449779,'s': 20.565774,'e':-64.493017,'w':-130.578836}
data = ult.random_points_extrema(1000000,extrema)
data
```

```python
# Geohashing the dataset 
data = ult.map_table(data,12)
data
```

```python
# Now placing the polygon index on top of the dataset we currently have
```

```python
data = ult.area_index(data,index)
data
```

```python
# Filtering the dataset for only points within an areaa in ultindex.
```

```python
data = data[data.AREA.astype(str).str.len() > 0]
```

```python
data
```

```python
data = mk.make_colorkey(data,'AREA')
```
